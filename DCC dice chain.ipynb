# Examine the math and odds of different rolls in DCC

## Spellcasting

what are the expected values for spellcheck rolls for different dies on the dice chain?


```python
import pandas as pd
```

```python
def spellcheck_mean(die:int, level:int):
    #what is the mean value for a spellcheck with a given die for a caster with the given level, and no other modifiers?
    rolls = [i + level +1 for i in range(die)]

    #add the caster level again for a crit
    rolls[-1]+=level

    return sum(rolls)/die
```

What is the expected value for a spell with the mercurial magic: whimsical patron?  

page 113, no 38: 
> Whimsical patron. The spell draws upon the wizard’s patron to succeed, but this entity’s attentions are notoriously fickle. Whenever this spell is cast, roll 1d6 to determine what die is used whenever this spell is cast:
(1-2) d16; (3-4) d20; (5) d24; (6) d30. This d6 roll supersedes the wizard’s normal action die and the dice chain.

```python
#Let's make a list of the expected value vs. caster level.

chain = [16, 20, 24, 30]
prob = [2/6, 2/6, 1/6, 1/6]
levels = range(1,11)
whimsy_dict = {f'level {lev}':{f'd{d}':spellcheck_mean(d, lev) for d in chain} for lev in levels}
spell_chain = pd.DataFrame(whimsy_dict)
spell_chain.index.name = 'die'
spell_chain
```

Now we can get the expected value.

```python
spell_chain.apply(lambda x: sum(x*prob))
```

We can get the expected value for a normal spellcheck from the d20 row in the spell_chain table. It's clear that the meurcural increases the expected value by 0.5.

```python
#make the table easier to read:

spell_chain.map(lambda x:'{:2.2f}'.format(x))
```
