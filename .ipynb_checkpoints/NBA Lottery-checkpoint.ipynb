# What are the odds that the NBA rigged its own lottery?

## Overview

This year (2025), for the fourth time in 40 years, a team with less than a 2% chance won the NBA's draft lottery. Anytime this happens, there is speculation about the lottery being rigged. For example, this year the Dallas Mavericks won it after having made what's widely expected to be one of the worst trades in league history, sending young superstar Luka Doncic to the Los Angeles Lakers in exchange for Anthony Davis, an ageing star with his best years behind him. Could the league have orchestrated the trade to get their most valuable team a superstar, assuring Dallas' management that they would get the number one pick in exchange?

It's a bit of an absurd conspiracy theory, and we can't determine its veracity using math alone, but we can at least calculate the odds of the lottery yielding so many low percentage winners in the time it's been around. This is the subject of this study.
