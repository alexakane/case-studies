What is the sum of all Fibonacci numbers smaller than or equal to 11112025? (required)

From Wikipedia:

https://en.wikipedia.org/wiki/Fibonacci_sequence

The Fibonacci sequence is a sequence in which each element is the sum of the two elements that precede it, with the first two being 0 and 1

```python
i = 0
j = 1
k = 0
fib = [i,j]
max = 11112025

while(i+j<=max):
    k = i+j
    i = j
    j = k
    fib.append(k)

fib
```

```python
print(fib[-1]<=max, fib[-1]+fib[-2]<=max)
```

```python
sum(fib)
```
