# The problem

A general problem that can arise in the human or medical service industries is, given the subpopulation of the community served by a specified service provider, what fractions of other subpopulations might exist in the served peopulation. For example, imagine that a local government body contracts with a nonprofit to provide health services for Medicare patients, and the government body knows the fraction of those patients that belong to a specific subplan, what is the number of patients served by the nonprofit that belong to that subplan?  

The only way to properly answer this question is to get the data directly, through polling, or using the PHI from the government, but in some cases those options may not be available, or be prohibitively expensive. Instead, here, we address a related question that may (or not) still provide useful information for decision makers, that is, if the nonprofit's clientele are *randomly selected* from the Medicare patients, what is the probability distribution of the number of clients belonging to the subplan?

## Statistics of the subpopulation

This question is readily answered using basic probability and statistics, as it is a case of choosing items from a population of two types without replacement. (In this case the two types are patients belonging to the subplan, and patients not belonging to it.)
